A2. Create a linear regression model on this data. Consider using one feature as independent 
variable while the other as dependent variable (you may also round this number to integer). After 
the model is created, calculate the mean square error by predicting the values from the model.

```python

#A2
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Sample data (replace with your own data)
df = pd.read_excel("D:\embeddingsdata.xlsx")

 
#Load features into variables  
df1=np.array(df['embed_8']).reshape(-1,1)
df2=np.array(df['embed_9'])

# Create a linear regression model
model = LinearRegression()
model.fit(df1,df2)

# Predict values using the model
df2_pred = model.predict(df1)

# Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(df2,df2_pred)

print(f'Linear Regression Model Coefficients: Intercept={model.intercept_}, Slope={model.coef_[0]}')
print(f'Mean Squared Error (MSE): {mse}')
```

```python
#A3 Study all the parameters and attributes associated with linear regression model. 
from sklearn.linear_model import LinearRegression

# Sample data (replace with your own data)
df1=np.array(df['embed_8']).reshape(-1,1)
df2=np.array(df['embed_9'])

# Create a linear regression model
model = LinearRegression()
model.fit(df1,df2)

# Accessing model parameters and attributes
intercept = model.intercept_
coefficients = model.coef_
rank = model.rank_
singular_values = model.singular_
copy_X = model.copy_X
n_features_in = model.n_features_in_

# Printing the results



print('The obtained results on attributes and parameters on regression model :')

print("Intercept:", intercept)
print("Coefficients (Slopes):", coefficients)
print("Effective Rank of Features:", rank)
print("Singular Values of Features:", singular_values)
print("Copy X during Fitting:", copy_X)
print("Number of Features Seen During Fitting:", n_features_in)

# Accessing and printing hyperparameters
hyperparameters = model.get_params()
print("\nHyperparameters:")
for key, value in hyperparameters.items():
    print(f"{key}: {value}")


```
